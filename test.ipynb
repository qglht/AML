```python
import pandas as pd
```

```python
path_x_train = "data/X_train.csv"
path_y_train = "data/y_train.csv"
```

```python
x_train, y_train = pd.read_csv(path_x_train), pd.read_csv(path_y_train)
```

```python
x_train
```

```python
x_train.describe()
```

```python
for column in x_train.columns:
    x_train[column] = x_train[column].fillna(x_train[column].median())
```

```python
x_train = (x_train-x_train.mean())/x_train.std()
for column in x_train.columns:
    x_train[column] = x_train[column].fillna(x_train[column].median())
```

```python
x_train
```

```python
pd.options.plotting.backend = "plotly"
column = "x1"

df = pd.DataFrame(dict(a=x_train["id"].values, b=x_train[column].values))
fig = df.plot()
fig.show()
```

```python
for column in df.columns:
    if df[column].isna().sum():
        print(column)
x_train.describe()
```

```python
import pandas
from sklearn.decomposition import PCA
import numpy
import matplotlib.pyplot as plot

df = x_train
df_normalized = x_train

pca = PCA(n_components=10)
pca.fit(df_normalized)

# Reformat and view results
loadings = pandas.DataFrame(pca.components_.T,
columns=['PC%s' % _ for _ in range(len(df_normalized.columns))],
index=df.columns)
print(loadings)

plot.plot(pca.explained_variance_ratio_)
plot.ylabel('Explained Variance')
plot.xlabel('Components')
plot.show()
```
